```python
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
import timeit
from IPython.display import clear_output
```

```python
# create sample data
n = 3000
x = np.random.rand(n)
y = np.random.rand(n)
```

```python
# create plot
start = timeit.default_timer()
fig = px.scatter(x=x,y=y,render_mode='svg')
fig.show()
end = timeit.default_timer()
print(end-start)
```

```python
start = timeit.default_timer()
fig = go.Figure(data=go.Scattergl(x=x,y=y,mode='markers'))
#fig.show()
end = timeit.default_timer()
print(end-start)
```

```python
times = np.arange(1,800000,10000)
measures_svg = []
measures_gl = []

for i in times:
    x = np.random.rand(i)
    y = np.random.rand(i)
    start = timeit.default_timer()
    fig = px.scatter(x=x,y=y,render_mode='svg')
    fig.show()
    clear_output()
    end = timeit.default_timer()
    measures_svg.append(end-start)

for i in times:
    x = np.random.rand(i)
    y = np.random.rand(i)
    start = timeit.default_timer()
    fig = go.Figure(data=go.Scattergl(x=x,y=y,mode='markers'))
    fig.show()
    clear_output()
    end = timeit.default_timer()
    measures_gl.append(end-start)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(times, measures_svg, color="green", label='svg')
ax.plot(times, measures_gl, color="red", label='gl')
plt.legend()
plt.xlabel('nr. of datapoints')
plt.ylabel('time in seconds')
plt.title('SVG vs WebGL rendering times')
plt.show()
```

```python
N = 5000

# Create figure
fig = go.Figure()
start = timeit.default_timer()
fig.add_trace(
    go.Scatter(
        x = np.random.randn(N),
        y = np.random.randn(N),
        mode = 'markers',
        marker = dict(
            line = dict(
                width = 1,
                color = 'DarkSlateGrey')
        )
    )
)

fig.show()
end = timeit.default_timer()
print(end-start)
```

```python


times = np.arange(1,800000,10000)
measures = []

for i in times:
    
    fig = go.Figure()
    start = timeit.default_timer()
    fig.add_trace(
        go.Scatter(
            x = np.random.rand(i),
            y = np.random.rand(i),
            mode = 'markers',
            marker = dict(
                line = dict(
                    width = 1,
                    color = 'DarkSlateGrey')
            )
        )
    )
    fig.show()
    clear_output()
    end = timeit.default_timer()
    measures.append(end-start)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(times, measures, color="green", label='svg')
plt.legend()
plt.xlabel('nr. of datapoints')
plt.ylabel('time in seconds')
plt.title('SVG vs WebGL rendering times')
plt.show()
```
